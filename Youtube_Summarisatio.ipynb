```python
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
```

```python
def get_transcript(video_url):
    try:
        video_id = video_url.split("v=")[1]
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        text = " ".join([t['text'] for t in transcript])
        return text
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
def summarize_text(text, sentences_count=5):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    stemmer = Stemmer("english")
    summarizer = LsaSummarizer(stemmer)
    summarizer.stop_words = get_stop_words("english")
    summary = summarizer(parser.document, sentences_count)
    return " ".join([str(sentence) for sentence in summary])
```

```python
def summarize_youtube_video(video_url, sentences_count=5):
    # Step 1: Get the transcript
    transcript = get_transcript(video_url)
    if transcript.startswith("Error"):
        return transcript

    # Step 2: Summarize the transcript
    summary = summarize_text(transcript, sentences_count)
    return summary
```

```python
def chunk_text(text, chunk_size=512):
    words = text.split()
    return [" ".join(words[i:i + chunk_size]) for i in range(0, len(words), chunk_size)]

from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_large_text(text):
    chunks = chunk_text(text)
    summaries = [summarizer(chunk, max_length=100, min_length=30, do_sample=False)[0]['summary_text'] for chunk in chunks]
    return " ".join(summaries)

summary_english = summarize_large_text(english_transcript)
summary_hindi = summarize_large_text(hindi_transcript)

print("English Summary:", summary_english)
print("Hindi Summary:", summary_hindi)
```

```python
# Install required libraries
!pip install yt-dlp youtube-transcript-api google-generativeai

# Import libraries
from yt_dlp import YoutubeDL
from youtube_transcript_api import YouTubeTranscriptApi
import google.generativeai as genai
import re

# Set up Gemini API
GOOGLE_API_KEY = "YOUR_GOOGLE_API_KEY"  # Replace with your Gemini API key
genai.configure(api_key=GOOGLE_API_KEY)

# Initialize Gemini model
model = genai.GenerativeModel('gemini-pro')

# Function to extract video ID from URL
def extract_video_id(url):
    # Handle different YouTube URL formats
    regex = r"(?:v=|\/)([0-9A-Za-z_-]{11}).*"
    match = re.search(regex, url)
    if match:
        return match.group(1)
    else:
        raise ValueError("Invalid YouTube URL. Could not extract video ID.")

# Function to extract video metadata using yt-dlp
def extract_metadata(url):
    ydl_opts = {}
    with YoutubeDL(ydl_opts) as ydl:
        try:
            info_dict = ydl.extract_info(url, download=False)
            metadata = {
                "title": info_dict.get("title", "No Title"),
                "description": info_dict.get("description", "No Description"),
                "views": info_dict.get("view_count", "N/A"),
                "length": info_dict.get("duration", "N/A"),
                "author": info_dict.get("uploader", "N/A"),
                "publish_date": info_dict.get("upload_date", "N/A")
            }
            return metadata
        except Exception as e:
            print(f"Error extracting metadata: {e}")
            return None

# Function to get transcript
def get_transcript(video_id):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        full_text = " ".join([entry['text'] for entry in transcript])
        return full_text
    except Exception as e:
        print(f"Error fetching transcript: {e}")
        return None

# Function to generate summary using Gemini
def generate_summary(text):
    try:
        response = model.generate_content(f"Summarize the following text in 3-4 sentences:\n{text}")
        return response.text
    except Exception as e:
        print(f"Error generating summary: {e}")
        return None

# Main pipeline
def video_summary_pipeline(url):
    try:
        # Step 1: Extract video ID
        video_id = extract_video_id(url)
        print("Video ID:", video_id)

        # Step 2: Extract metadata
        metadata = extract_metadata(url)
        if not metadata:
            print("Failed to extract metadata. Skipping summary generation.")
            return

        print("\nTitle:", metadata["title"])
        print("Description:", metadata["description"])
        print("Author:", metadata["author"])
        print("Length (seconds):", metadata["length"])
        print("Publish Date:", metadata["publish_date"])
        print("Views:", metadata["views"])

        # Step 3: Get transcript
        transcript = get_transcript(video_id)
        if not transcript:
            print("\nNo transcript available for this video.")
            return

        # Step 4: Generate summary using Gemini
        summary = generate_summary(transcript)
        if not summary:
            print("\nFailed to generate summary.")
            return

        print("\nVideo Summary:")
        print(summary)

    except Exception as e:
        print(f"An error occurred: {e}")

# Input YouTube URL
video_url = input("Enter the YouTube video URL: ")
video_summary_pipeline(video_url)
```
